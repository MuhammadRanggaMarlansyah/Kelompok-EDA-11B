###**Soal Nomor 1 dan 2**

1. Apakah ada dosen yang cenderung memberikan nilai cenderung lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya?

2. Menurut data yang ada, apakah terdapat perbedaan dalam pemberian nilai antara dosen dengan Pendidikan Terakhir S3 dan S2?

```python
nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]

stdx = nilai.groupby(nilai['NIDN']).std()[['nilai']]
stdx = stdx.rename({'nilai': 'std_nilai'}, axis=1)

nilai_dosen = nilai_dosen.set_index(['NIDN'])
nilai_dosen = pd.concat([nilai_dosen, stdx], axis = 1)
nilai_dosen
```

####Interpretasi No. 1 : Kecenderungan Pemberian Nilai oleh Dosen
Berdasarkan tabel diatas dengan mengamati standar deviasi dan rata-rata nilai, kami dapat mengetahui bahwa Pak  Andri Basuki adalah dosen yang memberikan nilai lebih tinggi kepada mahasiswa sedangkan Ibu Annisa Fitria adalah dosen yang memberikan nilai yang lebih rendah. Namun interpretasi tersebut tidak dapat dibuktikan secara akurat, karena Pak Andri Basuki memasukkan nilai mahasiswa di seluruh kelas sedangkan Bu Annisa Fitria hanya memasukkan nilai dari satu kelas.



####Interpretasi No. 2 : Pengaruh Tingkat Pendidikan Dosen pada Pemberian Nilai
Dengan mengamati tingkat pendidikan terakhir, standar deviasi, dan rata-rata nilai dari keempat dosen yang sudah kami analisis, kami dapat menyimpulkan bahwa dosen dengan tingkat pendidikan akhir S3 yaitu Pak Andri Basuki dan Pak Zulfikar memberikan nilai lebih tinggi sedangkan dosen dengan tingkat pendidikan akhir S2 yaitu Bu Annisa Fitria dan Pak Yudi memberikan nilai lebih rendah. Sehingga menurut kami terdapat perbedaan dalam pemberian nilai antara dosen dengan Pendidikan Terakhir S3 dan S2. Namun, interpretasi tersebut tidak dapat dibuktikan secara akurat, karena terdapat perbedaan jumlah kelas yang dinilai setiap dosen.

###**Soal Nomor 3**

Apakah terdapat perbedaan performa akademik antara mahasiswa yang berasal dari Madrasah/SMK dan SMU?

```python
asal_mahasiswa = mahasiswa[['nim','nama', 'Asal_Sekolah']]
nilai_mahasiswa = nilai[['nim', 'nilai']]
data_gabung = pd.merge(asal_mahasiswa, nilai_mahasiswa, on = 'nim', how = 'left')
data_gabung = data_gabung.set_index(['Asal_Sekolah', 'nim', 'nama'])
data_gabung = data_gabung.groupby(['Asal_Sekolah'])[['nilai']].mean()
data_gabung = data_gabung.rename({'nilai': 'mean_nilai'}, axis = 1)
data_gabung
```

####Interpretasi No. 3 : Hasil Nilai Berdasarkan Asal Sekolah Mahasiswa
Berdasarkan hasil analisis data di atas dapat kita ketahui bahwa rata-rata nilai mahasiswa yang berasal dari SMK lebih tinggi dibandingkan dengan rata-rata nilai mahasiswa yang berasal dari SMU maupun MA. Dalam kasus ini kita dapat simpulkan bahwa kemampuan akademik mahasiswa yang berasal dari SMK lebih tinggi dibandingkan mahasiswa yang berasal dari SMU maupun MA. Hal tersebut tentunya menarik, karena mengingat mata kuliah yang diberikan adalah ilmu matematika (murni) yang notabenenya anak SMU seharusnya lebih mengerti mengenai ilmu tersebut, dan juga ada kemungkinan bahwa mahasiswa yang berasal dari SMK lebih memahami tujuan dan tidak salah jurusan, sehingga mereka dapat sungguh-sungguh dalam mengikuti perkuliahan. Sedangkan mahasiswa yang berasal dari SMU/MA ada kemungkinan lintas jurusan atau faktor lainnya sehingga dapat mempengaruhi aktivitas perkuliahan mereka. Jadi asal sekolah tidak terlalu mempengaruhi kemampuan akademik seseorang. Kami merekomendasikan kepada perusahaan untuk membuat suatu program bimbingan yang lebih intensif kepada siswa SMU maupun MA, supaya para siswa tersebut dapat bersaing dalam dunia perkuliahan nanti.

Sedemikian sehingga dapat disimpulkan menurut hasil analisa di atas, rerata nilai akademik antara mahasiswa angkatan 2019 dan mahasiswa angkatan 2020 memiliki perbedaan yang signifikan.
